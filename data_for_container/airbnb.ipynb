```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/listings_54_columns.csv")
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
df = df.loc[~(df['price'] == '')]
df.index = pd.RangeIndex(len(df))
df["price"] = df["price"].str.slice(start=1, stop=None)  # Corta o $ da frente.
df["price"] = df["price"].str.slice(start=None, stop=-3) # Corta os .00 de tras.
df["price"] = df["price"].str.replace(',', '') # Apaga a virgule.
df['price'] = df['price'].astype(int, errors='raise')
```

```python
df['security_deposit'] = df['security_deposit'].replace(np.nan, 0)
df['cleaning_fee'] = df['cleaning_fee'].replace(np.nan, 0)
df['extra_people'] = df['extra_people'].replace(np.nan, 0)
df["security_deposit"] = df["security_deposit"].str.slice(start=1, stop=None)
df['security_deposit'] = df['security_deposit'].replace(np.nan, 0)
df["security_deposit"] = df["security_deposit"].str.slice(start=None, stop=-3)
df['security_deposit'] = df['security_deposit'].replace(np.nan, 0)
df["security_deposit"] = df["security_deposit"].str.replace(',', '')
df['security_deposit'] = df['security_deposit'].replace(np.nan, 0)
df['security_deposit'] = df['security_deposit'].astype(int, errors='raise')
df["cleaning_fee"] = df["cleaning_fee"].str.slice(start=1, stop=-3)
df["cleaning_fee"] = df["cleaning_fee"].str.replace(',', '')
df['cleaning_fee'] = df['cleaning_fee'].replace(np.nan, 0)
df['cleaning_fee'] = df['cleaning_fee'].astype(int, errors='raise')
df["extra_people"] = df["extra_people"].str.slice(start=1, stop=-3)
df["extra_people"] = df["extra_people"].str.replace(',', '')
df['extra_people'] = df['extra_people'].astype(int, errors='raise')
df = df.drop(columns=['host_acceptance_rate', 'square_feet'])
```
